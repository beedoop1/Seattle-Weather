# Removing unnecessary columns from both datasets

# Looking at the datasets and then selecting one station from each dataset

```python
# defining a value of 1461 as that would account for every day over the course of 4 years,
# so we can see the precipitation for every day
N_keep = 1461
df_seattle_subsample = df_seattle.sample(N_keep)
```

```python
df_seattle_subsample.head(10)
```

```python
# sorting the dataframe by date
df_seattle_subsample = df_seattle_subsample.sort_values(by="DATE")
```

```python
df_seatac = df_seattle.loc[df_seattle["NAME"] == "SEATTLE TACOMA AIRPORT, WA US"]
df_jfk = df_ny.loc[df_ny["NAME"] == "JFK INTERNATIONAL AIRPORT, NY US"]
```

```python
# checking to see if we are missing any values or have more
# seatac station missing values
# jfk station has an extra values
print(df_seatac.shape)
print(df_jfk.shape)
```

# Merging and tidying datasets to show only relevant columns

```python
df_jfk = df_jfk.drop_duplicates(subset="DATE")
df_seatac = df_seatac.drop_duplicates(subset="DATE")
```

```python
# checking again to see if we are missing any values or have more
# seatac station still missing values
# jfk station has correct amount of values
print(df_seatac.shape)
print(df_jfk.shape)
```

```python
df = df_jfk.merge(df_seatac, on="DATE", how="left")
```

```python
df.tail()
```

```python
# removing any duplicate dates before merging only the date and precipitation column of the two dataframes
df_jfk = df_jfk.drop_duplicates(subset="DATE")
df_seatac = df_seatac.drop_duplicates(subset="DATE")
df = df_jfk[['DATE', 'PRCP']].merge(df_seatac[['DATE', 'PRCP']], on='DATE', how='left')
```

```python
df.shape
```

```python
# # tidying the data by using melt to make our data longer
# df_tidy = pd.melt(df, id_vars="DATE", value_vars="PRCP", var_name="city")
```

```python
# df_tidy
```

```python
df = pd.melt(df, id_vars = 'DATE', var_name = 'CITY', value_name = 'PRCP')
```

```python
df.head()
```

# Renaming the elements and column names

```python
df.loc[df['CITY']=='PRCP_x', 'CITY'] = 'NYC'
df.loc[df['CITY']=='PRCP_y', 'CITY'] = 'SEA'
```

```python
df.head()
```

```python
df.tail()
```

```python
# creating a dictionary to help rename our columns
mapping_dict = {'DATE':'date', 'CITY':'city', 'PRCP':'precipitation'}
```

```python
df = df.rename(columns=mapping_dict)
```

```python
df.head()
```

# Creating derived variables and dropping rows with missing values

```python
# imputing the missing values using linear interpolation
df["precip_interp_1"] = df["precipitation"].interpolate()
```

```python
df["precip_interp_1"].isna().sum()
```

```python
df["precipitation"].isna().sum()
```

```python
# creating new column
# taking the average value on equivalent day-of-year from other years
df["precip_interp_2"] = df["precipitation"]
df_sea = df.loc[df["city"] == "SEA"]
df_sea["day_of_year"] = df_sea["date"].dt.dayofyear
nan_idx = pd.isna(df_sea["precipitation"])
df_mean_daily = df_sea.groupby("day_of_year")["precipitation"].mean()
```

```python
df_sea.loc[nan_idx, "precip_interp_2"] = df_mean_daily.loc[df_sea.loc[nan_idx, "day_of_year"].values].values
```

```python
df_sea["precip_interp_2"].isna().sum()
```

```python
df.loc[df_sea.index] = df_sea
```

```python
df.tail()
```

```python
# dropping the rows with a missing value
df = df.dropna(subset=["precipitation"])
df.info()
```
