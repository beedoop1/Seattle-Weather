#### Making new dataframes so that it is easier to compare the amount of precipitation from both cities across 2022 and 2023.

```python
df_sea = df[df['city'] == 'SEA']
df_ny = df[df['city'] == 'NYC']
```

```python
df_sea_2022 = df_sea[df_sea['year'] == 2022]
df_ny_2022 = df_ny[df_ny['year'] == 2022]

df_sea_2023 = df_sea[df_sea['year'] == 2023]
df_ny_2023 = df_ny[df_ny['year'] == 2023]
```

#### Making a scatterplot and line chart comparing the precipiation in both cities in 2022

```python
plt.figure(figsize=(10, 5))
plt.scatter(df_sea_2022['date'], df_sea_2022['precipitation'])
plt.scatter(df_ny_2022['date'], df_ny_2022['precipitation'], color = "orange")
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Precipitation (mm)', fontsize = 15)
plt.title('Precipitation in Seattle and New York 2022')
plt.legend(["Seattle", "New York"])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_sea_2022['date'], df_sea_2022['precipitation'])
plt.plot(df_ny_2022['date'], df_ny_2022['precipitation'], color = 'orange')
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Precipitation (mm)', fontsize = 15)
plt.title('Precipitation in Seattle and New York 2022')
plt.legend(["Seattle", "New York"])
```
